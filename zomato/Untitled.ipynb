```python
#What type of restaurant does the majority of customers order from?

restaurant_type_counts = df['listed_in(type)'].value_counts()
plt.figure(figsize=(10, 6))
restaurant_type_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Restaurant Types', fontsize=16)
plt.xlabel('Type of Restaurant', fontsize=12)
plt.ylabel('Number of Restaurants', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
 #How many votes has each type of restaurant received from customers?

votes_by_type = df.groupby('listed_in(type)')['votes'].sum().sort_values(ascending=False)

# Plotting the data as a pie chart
plt.figure(figsize=(8, 8))
votes_by_type.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors,
    textprops={'fontsize': 10}
)

# Adding title
plt.title('Vote Distribution by Restaurant Type', fontsize=14)

# Adding legend
plt.legend(votes_by_type.index, title='Restaurant Types', bbox_to_anchor=(1.05, 1), loc='upper left')

# Display the plot
plt.ylabel('')  # Remove the default y-axis label
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['rate'], bins=20, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribution of Restaurant Ratings', fontsize=16)
plt.xlabel('Ratings', fontsize=14)
plt.ylabel('Number of Restaurants', fontsize=14)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
online_orders = df[df['online_order'] == 'Yes']

# Calculate average spending for couples who order online
average_spending = online_orders['approx_cost(for two people)'].mean()

# Prepare data for plotting
categories = ['Average Spending']
values = [average_spending]

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='skyblue')
plt.title('Average Spending of Couples on Online Food Orders')
plt.ylabel('Average Cost (in currency)')
plt.ylim(0, max(values) + 100)  # Set y-axis limit slightly higher than max value
```

```python
 #Which mode (online or offline) has received the maximum rating?
average_ratings = df.groupby('online_order')['rate'].mean()

# Prepare data for plotting
modes = average_ratings.index
ratings = average_ratings.values

# Plotting the average ratings by mode
plt.figure(figsize=(8, 5))
plt.bar(modes, ratings, color=['skyblue', 'lightcoral'])
plt.title('Average Ratings by Order Mode (Online vs Offline)')
plt.xlabel('Order Mode')
plt.ylabel('Average Rating (out of 5)')
plt.ylim(0, 5)  # Set y-axis limit from 0 to 5
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Which type of restaurant received more offline orders, so that Zomato can provide those customers with some good offers?
offline_orders = df[df['online_order'] == 'No']

# Count the number of offline orders by restaurant type
offline_counts = offline_orders['listed_in(type)'].value_counts()

# Prepare data for plotting
restaurant_types = offline_counts.index
order_counts = offline_counts.values

# Plotting the number of offline orders by restaurant type
plt.figure(figsize=(12, 6))
plt.bar(restaurant_types, order_counts, color='cornflowerblue')
plt.title('Number of Offline Orders by Restaurant Type')
plt.xlabel('Restaurant Type')
plt.ylabel('Number of Offline Orders')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```
